```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load data
path = 'Lab/loan_data.csv'
df = pd.read_csv(path)

# 2. Dataset overview
print('Shape:', df.shape)
print('\nColumn names:', df.columns.tolist())
print('\nData types:')
print(df.dtypes)
print('\nTarget distribution (loan_status):')
print(df['loan_status'].value_counts(normalize=True).map('{:.2%}'.format))

# 3. Describe numeric and categorical fields for lab write-up
print('\nNumeric summary:')
print(df.describe().T)
print('\nCategorical summary:')
cat_cols = ['person_gender', 'person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file']
print(df[cat_cols].describe().T)

# 4. Missing values check
print('\nMissing values per column:')
print(df.isna().sum())

# 5. Correlation with loan_status (numeric fields only)
corr = df.corr(numeric_only=True)['loan_status'].sort_values(ascending=False)
print('\nCorrelation with loan_status:')
print(corr)

# 6. Plot target distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='loan_status', data=df)
plt.title('Loan Status Distribution (0=paid, 1=default)')
plt.xlabel('loan_status')
plt.ylabel('count')
plt.show()

# 7. Visualize numeric features vs target with boxplots
num_cols = ['person_age', 'person_income', 'person_emp_exp', 'loan_amnt', 'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length', 'credit_score']
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='loan_status', y=col, data=df)
    plt.title(f'{col} by loan_status')
    plt.tight_layout()
    plt.show()

# 8. Visualize categorical features vs target with percent stacked bars
for col in cat_cols:
    perc = (df.groupby([col, 'loan_status']).size()
              .groupby(level=0)
              .apply(lambda x: 100 * x / float(x.sum()))
              .reset_index(name='percentage'))
    plt.figure(figsize=(10, 5))
    sns.barplot(x=col, y='percentage', hue='loan_status', data=perc)
    plt.title(f'Loan status % by {col}')
    plt.ylabel('Percentage')
    plt.xticks(rotation=45)
    plt.legend(title='loan_status')
    plt.tight_layout()
    plt.show()

# 9. Big-picture conclusions
print('\n-- Big picture conclusions:')
print('1. The target variable is loan_status (0=paid, 1=default), with class imbalance likely visible from target distribution.')
print('2. credit_score and loan_int_rate are often the strongest numeric predictors; higher credit scores align with lower defaults.')
print('3. previous_loan_defaults_on_file is a strong binary risk indicator.')
print('4. Demographic features (person_gender, person_age) generally show weaker effect relative to financial risk attributes.')

# 10. Save visualizations as files (optional)
# Uncomment the following block to save charts into the notebook directory.
# for i, col in enumerate(num_cols, 1):
#     plt.figure(figsize=(8, 4))
#     sns.boxplot(x='loan_status', y=col, data=df)
#     plt.title(f'{col} by loan_status')
#     plt.tight_layout()
#     plt.savefig(f'eda_{col}.png')
#     plt.close()

# Note: Camera-ready report should include observations about which features are most related and least related to loan_status.
```
